```python
# Fractional Knapsack Problem using Greedy Approach

def fractional_knapsack(W, weights, values):
    # Calculate value/weight ratio
    ratio = [(values[i] / weights[i], weights[i], values[i]) for i in range(len(values))]
    # Sort by ratio in descending order
    ratio.sort(reverse=True)

    total_value = 0.0

    for r, wt, val in ratio:
        if W >= wt:
            W -= wt
            total_value += val
        else:
            total_value += r * W
            break

    return total_value

# Example usage
values = [60, 100, 120]
weights = [10, 20, 30]
W = 50

max_value = fractional_knapsack(W, weights, values)
print("Maximum value in Knapsack =", max_value)
```
