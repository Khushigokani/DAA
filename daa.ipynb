```python
#mergesort

def merge_sort(arr):
  if(len(arr)<=1):
     return arr
  mid=len(arr)//2
  left=arr[:mid]
  right=arr[mid:]
  left=merge_sort(left)
  right=merge_sort(right)
  return merged(left,right)

def merged(left,right):
  merge=[]
  i=j=0
  while i<len(left) and j<len(right):
    if(left[i]<right[j]):
      merge.append(left[i])
      i=i+1
    else:
      merge.append(right[j])
      j=j+1
  while(i<len(left)):
    merge.append(left[i])
    i=i+1
  while(j<len(right)):
    merge.append(right[j])
    j=j+1
  return merge

arr=[2,4,1,5,7,3,6]
print(merge_sort(arr))
```

```python
#binary search

def binarysearch(arr,l,r,key):
  mid=(l+r)//2
  if(r>=l):
    if(key==arr[mid]):
      return mid
    elif(key<arr[mid]):
      return binarysearch(arr,l,mid-1,key)
    else:
      return binarysearch(arr,mid+1,r, key)
  else:
    return-1

arr=[1,3,5,7,9,11,13,15]
key=11 #if placing 10 or 12 it shows -1
l=0
r=len(arr)-1
print(binarysearch(arr,l,r,key))
```

```python
#insertion sort

def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1

        while j >= 0 and arr[j] > key:
            arr[j + 1] = arr[j]
            j -= 1

        arr[j + 1] = key
arr = [5, 3, 8, 1, 2]
insertion_sort(arr)
print("Insertion Sort:", arr)
```

```python
#bubble sort

def bubble_sort(arr):
    n = len(arr)

    for i in range(n):
        for j in range(n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]

arr = [5, 3, 8, 1, 2]
bubble_sort(arr)
print("Bubble Sort:", arr)
```

```python
#selection sort

def selection_sort(arr):
    n = len(arr)

    for i in range(n):
        min_index = i

        for j in range(i + 1, n):
            if arr[j] < arr[min_index]:
                min_index = j

        arr[i], arr[min_index] = arr[min_index], arr[i]

arr = [5, 3, 8, 1, 2]
selection_sort(arr)
print("Selection Sort:", arr)
```

```python
#counting sort

def counting_sort(arr):
    if not arr:
        return

    max_val = max(arr)
    count = [0] * (max_val + 1)

    for num in arr:
        count[num] += 1

    i = 0
    for num in range(len(count)):
        while count[num] > 0:
            arr[i] = num
            i += 1
            count[num] -= 1

# Example
arr = [4, 2, 2, 8, 3, 3, 1]
counting_sort(arr)
print("Counting Sort:", arr)
```

```python
#quick sort
def partition(arr, low, high):
    pivot = arr[high]
    pIndex = low
    for j in range(low, high):
        if arr[j] <= pivot:
            arr[pIndex], arr[j] = arr[j], arr[pIndex]
            pIndex=pIndex+1
    arr[pIndex], arr[high] = arr[high], arr[pIndex]
    return pIndex

def quicksort(arr, low, high):
    if low < high:
        pi = partition(arr, low, high)
        quicksort(arr, low, pi - 1)
        quicksort(arr, pi+1, high)

arr=[10,7,1,8,9,6,5]
low=0
high=len(arr)-1
quicksort(arr, low, high)
print(arr)
```

```python

# Long Integer Multiplication

def multiply(num1, num2):
    # Handle zero case
    if num1 == "0" or num2 == "0":
        return "0"

    # Convert strings to lists of integers (reverse for easier multiplication)
    num1 = [int(d) for d in num1[::-1]]
    num2 = [int(d) for d in num2[::-1]]

    # Result can be at most len(num1) + len(num2) digits
    result = [0] * (len(num1) + len(num2))

    # Multiply each digit
    for i in range(len(num1)):
        for j in range(len(num2)):
            result[i + j] += num1[i] * num2[j]
            # Handle carry
            result[i + j + 1] += result[i + j] // 10
            result[i + j] %= 10

    # Remove leading zeros
    while len(result) > 1 and result[-1] == 0:
        result.pop()

    # Reverse back and join as string
    return ''.join(map(str, result[::-1]))


# Example usage
num1 = "987654321987654321"
num2 = "123456789123456789"
product = multiply(num1, num2)
print(product)
```

```python
#karatsuba

def karatsuba(x, y):
    if x < 10 or y < 10:
        return x * y
    m = max(len(str(x)), len(str(y))) // 2
    a = x // (10**m)
    b = x % (10**m)
    c = y // (10**m)
    d = y % (10**m)

    ac = karatsuba(a, c)
    bd = karatsuba(b, d)
    ad_bc = karatsuba(a + b, c + d) - ac - bd

    return ac * (10**(2 * m)) + ad_bc * (10**m) + bd

# Ex
x = 1234
y = 5678
result = karatsuba(x, y)
print(result)
```

```python
#D & C last approach Min-Max
def minmax(arr):
  if(len(arr)==1):
    return (arr[0], arr[0])
  m=int(len(arr)/2)
  left=arr[:m]
  right=arr[m: ]
  lmin, lmax= minmax(left)
  rmin, rmax = minmax(right)
  return(min(lmin,rmin),max(lmax,rmax))

# Example usage:
arr = [3, 1, 4, 1, 5, 9, 2, 6]
minimum, maximum = minmax(arr)
print(minimum, maximum)
```

```python
# Fractional Knapsack Problem using Greedy Approach

def fractional_knapsack(W, weights, values):
    # Calculate value/weight ratio
    ratio = [(values[i] / weights[i], weights[i], values[i]) for i in range(len(values))]
    # Sort by ratio in descending order
    ratio.sort(reverse=True)

    total_value = 0.0

    for r, wt, val in ratio:
        if W >= wt:
            W -= wt
            total_value += val
        else:
            total_value += r * W
            break

    return total_value

# Example usage
values = [60, 100, 120]
weights = [10, 20, 30]
W = 50

max_value = fractional_knapsack(W, weights, values)
print("Maximum value in Knapsack =", max_value)
```

```python
#job scheduling

def job_scheduling(jobs):
    jobs.sort(key=lambda x: x[2], reverse=True)
    max_deadline = max(job[1] for job in jobs)
    slot = [-1] * (max_deadline + 1)

    profit = 0
    for job in jobs:
        for t in range(job[1], 0, -1):
            if slot[t] == -1:
                slot[t] = job[0]
                profit += job[2]
                break
    return profit

jobs = [('a',2,100),('b',1,19),('c',2,27),('d',1,25),('e',3,15)]
print(job_scheduling(jobs))
```

```python
#activity selection

def activity_selection(start, finish):
    activity = sorted(zip(start, finish), key=lambda x: x[1])
    selected = [activity[0]]

    for i in range(1, len(activity)):
        if activity[i][0] >= selected[-1][1]:
            selected.append(activity[i])

    return selected

print(activity_selection([1,3,0,5,8,5],[2,4,6,7,9,9]))
```

```python
#lcs

def lcs(a, b):
    m, n = len(a), len(b)
    dp = [[0]*(n+1) for _ in range(m+1)]

    for i in range(m):
        for j in range(n):
            if a[i] == b[j]:
                dp[i+1][j+1] = dp[i][j] + 1
            else:
                dp[i+1][j+1] = max(dp[i][j+1], dp[i+1][j])

    return dp[m][n]

print(lcs("ABCBDAB", "BDCABA"))
```

```python
#naive search

def naive_search(text, pattern):
    for i in range(len(text) - len(pattern) + 1):
        match = True
        for j in range(len(pattern)):
            if text[i+j] != pattern[j]:
                match = False
                break
        if match:
            return i
    return -1

print(naive_search("AABAACAADAABAABA", "AABA"))
```

```python
#rabin karp

def rabin_karp(text, pattern, q=101):
    m, n = len(pattern), len(text)
    d = 256

    p = t = 0
    h = 1

    for i in range(m-1):
        h = (h * d) % q

    for i in range(m):
        p = (d*p + ord(pattern[i])) % q
        t = (d*t + ord(text[i])) % q

    for i in range(n-m + 1):
        if p == t:
            if text[i:i+m] == pattern:
                return i

        if i < n - m:
            t = (d*(t - ord(text[i])*h) + ord(text[i+m])) % q
            if t < 0:
                t += q

    return -1

print(rabin_karp("AABAACAADAABAABA", "AABA"))
```

```python
# N-Queen Problem

def is_safe(board, row, col, n):
    for i in range(row):
        if board[i] == col or \
           abs(board[i] - col) == abs(i - row):
            return False
    return True


def solve_nqueen(board, row, n):
    if row == n:
        print(board)
        return True

    for col in range(n):
        if is_safe(board, row, col, n):
            board[row] = col
            solve_nqueen(board, row + 1, n)


n = 4
board = [-1] * n
solve_nqueen(board, 0, n)
```
